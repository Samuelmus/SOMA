```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from datetime import datetime
import time
import pandas as pd 
import sys

tabela = pd.read_excel("forms_completo.xlsx")
tabela['Data'] = pd.to_datetime(tabela['Data'], dayfirst=True, errors='coerce').dt.strftime('%d/%m/%Y')

caixa_mapping = {
    'VL_UPANEMA E MOSSORÓ_ITAÚ [CONTA CORRENTE]': '3490',
    #COLOQUE OS CAIXAS NESSE FORMATO, PUXE VIA HTML
    
}

def ajustar_valor(valor_str):
    try:
        valor_str = str(valor_str).strip()
        valor_str = valor_str.replace(',', '.')
        valor_float = float(valor_str)
        if '.' in valor_str:
            decimal_part = valor_str.split('.')[-1]
            if len(decimal_part) == 1 or (len(decimal_part) > 1 and decimal_part[1] == '0'):
                valor_float *= 10
        return valor_float
    except ValueError:
        return float('nan') 

tabela['Valor'] = tabela['Valor'].apply(ajustar_valor)


navegador = webdriver.Chrome()
navegador.get('https://www.verbodavida.info/apps/index.php')
wait = WebDriverWait(navegador, 10)

usuario_input = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="login"]/div[1]/div/input')))
usuario_input.send_keys("") 
#COLOQUE O EMAIL

senha_input = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="login"]/div[2]/div/input')))
senha_input.send_keys("")
#COLOQUE A SENHA

login_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="login"]/div[3]/div[2]/button')))
login_button.click()

button_soma = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="2845"]')))
button_soma.click()
time.sleep(5)

linha = len(tabela)
print(linha)
cont = 0 


for i, cpf in enumerate(tabela["CPF"]):
    cont = cont + 1
    data = tabela.loc[i, "Data"]
    descricao = tabela.loc[i, "Descrição"]
    valor = tabela.loc[i, "Valor"]
    caixa = tabela.loc[i, "Caixa"]
    centro = tabela.loc[i, "Centro"]
    plano = tabela.loc[i, "Plano"]
    forma = tabela.loc[i, "Forma"]

    navegador.get('https://www.verbodavida.info/IVV/?mod=ivv&exec=entradas_saidas')
    wait.until(EC.url_to_be('https://www.verbodavida.info/IVV/?mod=ivv&exec=entradas_saidas'))
    navegador.get('https://www.verbodavida.info/IVV/?mod=ivv&exec=entradas_saidas_dados')
    wait.until(EC.url_to_be('https://www.verbodavida.info/IVV/?mod=ivv&exec=entradas_saidas_dados'))

    radio_element = navegador.find_element(By.XPATH, '//*[@id="exampleStandardForm"]/div[3]/div/div[2]/input')
    radio_element.click()
    time.sleep(3)

    select_element = WebDriverWait(navegador, 10).until(
        EC.visibility_of_element_located((By.ID, 'plano')))
    select = Select(select_element)
    select.select_by_visible_text(plano)
 

    select_custo = WebDriverWait(navegador, 10).until(
        EC.visibility_of_element_located((By.XPATH, '//*[@id="exampleStandardForm"]/div[7]/div/select')))
    select_2 = Select(select_custo)
    select_2.select_by_visible_text(centro)

    input_element = WebDriverWait(navegador, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="exampleStandardForm"]/div[9]/div/input')))
    input_element.click()
    input_element.send_keys(descricao)
    time.sleep(2)

    input_element_2 = WebDriverWait(navegador, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="exampleStandardForm"]/div[11]/div/div/input')))
    input_element_2.click()
    time.sleep(1)
    input_element_2.send_keys(data)
    time.sleep(1)

    input_element_3 = WebDriverWait(navegador, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="exampleStandardForm"]/div[12]/div/input')))
    input_element_3.clear()
    time.sleep(1)


    input_element_3.send_keys(valor)
    time.sleep(2)

    input_element_4 = WebDriverWait(navegador, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="exampleStandardForm"]/div[16]/div/select')))
    input_element_4.send_keys(forma)
    time.sleep(2)

    select_caixa = WebDriverWait(navegador, 10).until(
        EC.visibility_of_element_located((By.NAME, 'id_caixa_origem')))
    select_3 = Select(select_caixa)

    if caixa in caixa_mapping:
        select_3.select_by_value(caixa_mapping[caixa])
    else:
        print(f"Valor de caixa '{caixa}' não encontrado no mapeamento.")

    button_enviar = navegador.find_element(By.XPATH, '//*[@id="btn_salvar"]')
    button_enviar.click()
    time.sleep(2)

    if cont == linha:
        navegador.quit() 
        sys.exit()
```
